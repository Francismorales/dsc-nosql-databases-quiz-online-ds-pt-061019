## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


MongoDB is a distributed database (not all storage devices are attached to a common processor). Relational databases are not
network connected like distributed databases.
MongoDB stores data into flexible JSON like documents, whereas relational databases follow the table-row-column structure






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault-tolerance is the capability of a computer system or a network to deliver uninterurpted service, despite of one or more
of its components failing. 

An RDD is a partitioned collection of records. Each dataset in RDD is divided into logical partitions across many servers so
that they can be computed on different nodes of the cluster.

RDD lineage graph [DAG] allows RDDs to be fault-tolerant by allowing to recompute missing or damaged partitions due to node 
failures
